```python
from data_set import LeucorrheaDataset
import numpy as np
import time
```

```python
data_set = LeucorrheaDataset(train=False)
```

```python
ratios =np.array([0.54, 0.88, 1.17, 1.63, 2.3])
anchor_sizes = np.array([50, 96, 184, 273, 395])

classes =np.zeros([5,4])

num_each_class = np.zeros([7])

length = len(data_set)
for epoch in range(100):
    
    classes = classes*0
    loss = 0
    size_loss = 0
    start_time = time.time()
    for i in range(length):
        img,target = data_set[i]
        boxes = target['boxes'].numpy()
        for j in range(len(boxes)):
            box = boxes[j]
            width = box[2]-box[0]
            hight = box[3]-box[1]
            
            ratio = hight/width     
            distance = np.square(ratios-ratio)
            index = np.argmin(distance)
            loss+=distance[index]
            cp = classes[index]
            cp[0]+=ratio
            cp[1]+=1
            
            size = np.sqrt(width*hight)
            size_distance = np.square(anchor_sizes-size)
            arg_size = np.argmin(size_distance)
            size_loss +=size_distance[arg_size]
            size_cp = classes[arg_size]
            size_cp[2]+=size
            size_cp[3]+=1
        if epoch==0:
            label = target['labels']
            num_each_class[label-1]+=1
    time_elapsed = time.time()-start_time
    ratios = classes[:,0]/(classes[:,1]+1e-10)
    anchor_sizes = classes[:,2]/(classes[:,3]+1e-10)
   
    print('epoch:{}/100    ratio_loss:{:.3f}    anchor_loss:{}  time:{}'.format(epoch,loss,size_loss,time_elapsed))
    print('ratios:{}'.format(ratios))
    print('anchor_sizes:{}'.format(anchor_sizes))
    print('---------------------------------------------------------------')
print('num_each_class:{}'.format(num_each_class))
    
```
